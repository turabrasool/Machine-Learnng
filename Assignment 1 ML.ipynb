```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("/content/ex1data1.txt", header=None,names=["house area", "house price"])

X = data.iloc[:,0]
y = data.iloc[:,1]


def plot_data(X, y):
    plt.scatter(X, y, marker='o', c='b')
    plt.title('Scatter Plot of Data')
    plt.xlabel('House Area')
    plt.ylabel('House Price')
    plt.show()

plot_data(X, y)
```

# New Section

```python
def compute_cost(X, y, theta):
    m = len(y)
    h = X @ theta
    J = (1 / (2 * m)) * np.sum((h - y)**2)
    return J

def gradient_descent(X, y, theta, alpha, num_iterations):
    m = len(y)
    J_history = np.zeros(num_iterations)


    for i in range(num_iterations):
        h = X @ theta
        errors = h - y
        gradient = (1 / m) * (X.T @ errors)
        theta = theta - alpha * gradient
        J_history[i] = compute_cost(X, y, theta)


    return theta, J_history


X = np.column_stack((np.ones(len(X)), X))
theta = np.zeros(X.shape[1])

alpha = 0.003
num_iterations = 1000


theta, J_history = gradient_descent(X, y, theta, alpha, num_iterations)


def plot_data_with_optimized_line(X, y_old, y_new):
    plt.scatter(X, y_old, marker='o', c='b', label='Original Data')
    plt.plot(X, y_new, c='r', label='Optimized Line')
    plt.title('Linear Regression with Optimized Line')
    plt.xlabel('House Area')
    plt.ylabel('House Price')
    plt.legend()
    plt.show()

plot_data_with_optimized_line(X[:, 1], y, X.dot(theta))
```

# New Section

```python
def plot_theta_convergence (J_history):
  plt.plot(np.arange(1, num_iterations + 1), J_history, label='Cost Function')
  plt.xlabel('Number of Iterations')
  plt.ylabel('Cost')
  plt.legend()
  plt.show()

plot_theta_convergence (J_history)
```
